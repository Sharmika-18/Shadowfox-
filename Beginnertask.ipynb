Matplotlib examples...

```python
import matplotlib.pyplot as plt

df = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [2, 5, 7, 1, 6],
    "category": ["A", "B", "C", "A", "B"]
})
```

```python
plt.figure(figsize=(5,3))
plt.plot(df["x"], df["y"], marker="o", color="blue")
plt.title("Matplotlib - Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(5,3))
plt.scatter(df["x"], df["y"], color="red")
plt.title("Matplotlib - Scatter Plot")
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(5,3))
plt.bar(df["x"], df["y"], color="green")
plt.title("Matplotlib - Bar Plot")
plt.show()
```

```python
# Histogram
plt.figure(figsize=(5,3))
plt.hist(df["y"], bins=5, color="orange", edgecolor="black")
plt.title("Matplotlib - Histogram")
plt.show()
```

```python
# Pie Chart
plt.figure(figsize=(5,3))
sizes = [40, 30, 20, 10]
labels = ["A", "B", "C", "D"]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Matplotlib - Pie Chart")
plt.show()
```

Pandas Example...

```python
import pandas as pd
df = pd.DataFrame({
    "x": [6,5,7,4,3],
    "y": [3,5,7,2,1],
    "category": ["A", "B", "C", "A", "B"]
})
```

```python
# Line Plot
df.plot(x="x", y="y", kind="line", marker="o", title="Pandas - Line Plot")
```

```python
# Scatter Plot
df.plot(x="x", y="y", kind="scatter", title="Pandas - Scatter Plot", color="red")
```

```python
# Bar Plot
df.plot(x="x", y="y", kind="bar", title="Pandas - Bar Plot", color="green")
```

```python
# Histogram
df["y"].plot(kind="hist", bins=5, title="Pandas - Histogram", color="orange", edgecolor="black")
```

```python
# Pie Chart (using Series)
df["category"].value_counts().plot(kind="pie", autopct='%1.1f%%', title="Pandas - Pie Chart")
```
